# Investigating Netflix Movies
**DataCamp Project** | Exploratory Data Analysis on Netflix 1990s Movies

Perform exploratory data analysis on `netflix_data.csv` to understand movies from the 1990s decade.
- What was the most frequent movie duration in the 1990s?
- How many short action movies (< 90 min) were released in the 1990s?

```python
# Importing pandas and matplotlib
import pandas as pd
import matplotlib.pyplot as plt

# Read in the Netflix CSV as a DataFrame
netflix_df = pd.read_csv('netflix_data.csv')
```

```python
# Start coding here! Use as many cells as you like

# Filter movies from the 1990s
movies_90s = netflix_df[(netflix_df['type'] == 'Movie') & (netflix_df['release_year'] >= 1990) & (netflix_df['release_year'] < 2000)]

# Most frequent movie duration in the 1990s
duration = int(movies_90s['duration'].mode()[0])
print('Most frequent duration:', duration)

# Count short action movies (< 90 min) released in the 1990s
short_movie_count = int(movies_90s[(movies_90s['genre'] == 'Action') & (movies_90s['duration'] < 90)].shape[0])
print('Short action movie count:', short_movie_count)
```
